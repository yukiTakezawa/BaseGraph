```python
length = []
n_nodes_list = []

max_degree = 20

for n_nodes in tqdm(range(2, 1000)):
    if max(list(sympy.factorint(n_nodes))) > max_degree + 1:
        pass
    else:
        n_nodes_list.append(n_nodes)
        g = HyperHyperCube(n_nodes, max_degree=max_degree)
        length.append(len(g.w_list))
        
plt.plot(n_nodes_list, length, marker="o")
plt.plot(n_nodes_list, [2*math.log(n, max_degree+1) for n in n_nodes_list])
plt.plot(n_nodes_list, [math.log(n, max_degree+1) for n in n_nodes_list])
```
