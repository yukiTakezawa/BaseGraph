```python
import sys
sys.path.append('../')

%matplotlib inline
import matplotlib.pyplot as plt
from one_peer_deco_graph import *
```

```python
g = OnePeerDecoGraph(1)

n_nodes = 200
dp_table, _ = g.solve(n_nodes, 1)
```

```python
def cost(n):
    if math.log2(n) == int(math.log2(n)):
        return math.log2(n)
    
    for i in range(n):
        if n% (2**i) != 0:
            break
    m = n /(2**i)
    
    return 2 * math.floor(math.log2(m)) + 1 + i
```

```python
plt.plot(dp_table, color="red", lw=1)
plt.plot([n for n in range(1, n_nodes+1)], [cost(n) for n in range(1, n_nodes+1)], color="blue", lw=1)
plt.ylabel("length")
plt.xlabel("#nodes")
```
