```python
import sys
sys.path.append('../')

import torch
import random
import math
import scienceplots
from adic_graph import *

from tqdm import tqdm
```

```python
def upper_bound(n_nodes, max_degree):
    return 2 * math.log(n_nodes, max_degree+1) + 2
```

```python
plt.style.use(['science', 'notebook'])
```

```python
length = {}
simple_length = {}
```

```python
for max_degree in [1, 2, 3]:
    length[max_degree] = []
    simple_length[max_degree] = []
    for n_nodes in tqdm(range(1, 200)):
        g = AdicGraph(n_nodes, max_degree=max_degree)
        length[max_degree].append(len(g.w_list))
        
        g = SimpleAdicGraph(n_nodes, max_degree=max_degree)
        simple_length[max_degree].append(len(g.w_list))
```
